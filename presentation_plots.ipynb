```python
import matplotlib.pyplot as plt
import numpy as np

# Discrete action sets
gamma_vals = np.array([0, 1])
sigma_vals = np.array([0, 1])

# Construct grids
G, S = np.meshgrid(gamma_vals, sigma_vals)

# Define payoffs consistent with potential game property
B1 = np.array([
    [2, 30],
    [0, 8],
])  

B2 =  np.array([
    [2, 0],
    [30, 8],
]) 

phi = np.array([
    [24, 22],
    [22, 0],
])

# Plot discrete heatmaps
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

im1 = axs[0].imshow(B1, cmap="Blues", origin="lower")
axs[0].set_title("Player 1 Payoff $B_1(\\gamma, \\sigma)$")
axs[0].set_xticks(range(len(gamma_vals)))
axs[0].set_xticklabels(gamma_vals)
axs[0].set_yticks(range(len(sigma_vals)))
axs[0].set_yticklabels(sigma_vals)
fig.colorbar(im1, ax=axs[0])

im2 = axs[1].imshow(B2, cmap="Reds", origin="lower")
axs[1].set_title("Player 2 Payoff $B_2(\\gamma, \\sigma)$")
axs[1].set_xticks(range(len(gamma_vals)))
axs[1].set_xticklabels(gamma_vals)
axs[1].set_yticks(range(len(sigma_vals)))
axs[1].set_yticklabels(sigma_vals)
fig.colorbar(im2, ax=axs[1])

im0 = axs[2].imshow(phi, cmap="Greens", origin="lower")
axs[2].set_title("Potential Function $\\phi(\\gamma, \\sigma)$")
axs[2].set_xticks(range(len(gamma_vals)))
axs[2].set_xticklabels(gamma_vals)
axs[2].set_yticks(range(len(sigma_vals)))
axs[2].set_yticklabels(sigma_vals)
fig.colorbar(im0, ax=axs[2])

plt.tight_layout()
plt.show()
```
